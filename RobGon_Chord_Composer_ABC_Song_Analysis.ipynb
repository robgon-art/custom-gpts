```python
import os

def rename_folders(root_path):
    for root, dirs, files in os.walk(root_path, topdown=False):
        for name in dirs:
            if '[' in name or ']' in name:
                new_name = name.replace('[', '').replace(']', '')
                # Construct the full path to both the old and new directories
                old_dir = os.path.join(root, name)
                new_dir = os.path.join(root, new_name)

                # Rename the directory
                os.rename(old_dir, new_dir)
                print(f"Renamed '{old_dir}' to '{new_dir}'")

# Set the root path to your dataset directory
root_path = '/content/OpenEWLD/dataset'
rename_folders(root_path)
```

```python
import subprocess
import sys

def convert_mxl_to_text(song_file_path):
    """
    Converts an MXL file to text using xml2abc.py and returns text in ABC fomat.

    Args:
    - song_file_path (str): Path to the MXL file to be converted.

    Returns:
    - output (str): Extracted lyrics in ABC format.
    """
    # Define the path to xml2abc.py script
    xml2abc_script_path = "/content/xml2abc_145/xml2abc.py"

    output = None

    try:
        # Call xml2abc.py with subprocess to convert MXL to ABC notation
        output_bytes = subprocess.check_output(["python", xml2abc_script_path, song_file_path, "-x", "-n", "9999", "-d", "4"], timeout=10) # "-u", "-b", "4",
        output = output_bytes.decode("utf-8").strip()

    except Exception as e:
        # Handle exceptions, such as timeouts or subprocess errors
        print("Unexpected error:", str(e))

    # Return the extracted lyrics
    return output

# Example usage
song_file_path = "/content/Trouble_in_Mind_Blues.mxl"
abc_content = convert_mxl_to_text(song_file_path)
print(abc_content)
```

```python
import re
def simplify_abc(abc_content):
    """
    Simplification of ABC notation

    Args:
    - abc_content (str): The original ABC content as a string.

    Returns:
    - str: The simplified ABC content.
    """
    # Split the content into lines for processing
    lines = abc_content.split('\n')

    simplified_lines = []
    titles_seen = False
    last_chord = None

    for line in lines:
        if line.startswith('T:'):
            # print(line)
            if titles_seen:
              # print('titles seen')
              continue
            titles_seen = True
            # print("title added")
            simplified_lines.append(line)
            continue

        if line.startswith('%%score') or line.startswith('I:linebreak $'):
            continue

        if line.startswith('M:') or line.startswith('K:'):
            simplified_lines.append(line)
            continue

        if line.startswith('L:') or line.startswith('X:') or line.startswith('V:') or line.startswith('w:'):
            continue

        else:
            if line.startswith('w:') or not any(char.isdigit() for char in line):
                # This is a lyric line or a header/meta line, add it directly
                simplified_lines.append(line.replace('$', '').replace('%', ''))
            else:
                # This is a music notation line, process to keep only chords and bars
                chord_line = ""
                parts = line.split('|')
                for part in parts:
                    # Use regex to find all occurrences of text within double quotes
                    extracted_texts = re.findall(r'"([^"]*)"', part)

                    # Join the extracted pieces with spaces, removing the commas as specified in the results
                    results = ' '.join(text.replace(',', '') for text in extracted_texts)

                    chord_line += '|' + results
                simplified_lines.append(chord_line)

    # Join the simplified lines back into a single string
    simplified_abc_content = '\n'.join(simplified_lines)
    return simplified_abc_content

simplified_abc = simplify_abc(abc_content)
print(simplified_abc)
```

```python
import openai
api_key = 'your_openai_api_key'
client = openai.Client(api_key=api_key)

def get_genres_styles(abc_content):
  prompt = f"Given this song in ABC format, create lists for genres and styles.\n{abc_content}\nReturn with:\ngenres: <comma-separated list of genres>\nstyles: <comma-separated list of styles>"
  results = client.chat.completions.create(
    model="gpt-3.5-turbo-1106",
    messages=[{"role": "user", "content": prompt}])

  lists = results.choices[0].message.content
  genres = []
  styles = []
  for line in lists.split('\n'):
      if line.lower().startswith('genres:'):
          genres = line.split(':')[1].lower().strip()
      elif line.lower().startswith('styles:'):
          styles = line.split(':')[1].lower().strip()
  return genres, styles

genres, styles = get_genres_styles(simplified_abc)
print("genres:", genres)
print("styles:", styles)
```

```python
import pandas as pd

# Initialize a list to store song details
songs_details = []

for i, sf in enumerate(song_files):
    print(i, sf)
    song_file_path = sf
    abc_content = convert_mxl_to_text(song_file_path)
    if abc_content == None:
        continue
    simplified_abc = simplify_abc(abc_content)

    # Split the simplified ABC content into lines for processing
    lines = simplified_abc.split('\n')

    # Initialize a dictionary to hold the details of the current song
    song_dict = {'title': '', 'composer': '', 'meter': '', 'key': '', 'chords': '', 'genres': '', 'styles': ''}

    for line in lines:
        if line.startswith('T:'):
            song_dict['title'] = line[2:]
        elif line.startswith('C:'):
            song_dict['composer'] = line[2:]
        elif line.startswith('M:'):
            song_dict['meter'] = line[2:]
        elif line.startswith('K:'):
            song_dict['key'] = line[2:]
        else:
            # Assuming the remaining lines are chords, and you want to capture them all
            song_dict['chords'] += line

    genres, styles = get_genres_styles(simplified_abc)
    song_dict['genres'] = genres
    song_dict['styles'] = styles

    songs_details.append(song_dict)

    # if i >= 9:
    #     break

    # if i % 50 == 0:
    #     print(i)

# Create a DataFrame from the song details
df = pd.DataFrame(songs_details)

# Define the CSV file path where the DataFrame will be saved
csv_file_path = '/content/public_domain_songs.csv'

# Save the DataFrame to a CSV file
df.to_csv(csv_file_path, index=False)
```
